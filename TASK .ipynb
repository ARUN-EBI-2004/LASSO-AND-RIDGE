```python
import pandas as pd
```

```python
tour=pd.read_csv('Top Indian Places to Visit.csv')
```

```python
tour=tour[tour['Entrance Fee in INR']!=0]
```

```python
tour.shape
```

```python
tour
```

```python
x=tour[['Type','Google review rating','Number of google review in lakhs']]
y=tour['Entrance Fee in INR']
```

```python
x=pd.get_dummies(x,columns=['Type'],dtype=int)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
model1=Lasso(alpha=50,max_iter=150,tol=0.2)
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
model2=Ridge(alpha=100,max_iter=150,tol=0.2)
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```
