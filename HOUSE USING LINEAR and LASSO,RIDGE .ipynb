```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
house=pd.read_csv('Melbourne_housing_FULL.csv')
```

```python
house.head()
```

```python
house=house.drop(columns=['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'])
```

```python
house
```

```python
house.isnull().sum()
```

```python
house=house.dropna()
```

```python
house.nunique()
```

```python
house.shape
```

```python
x=house.drop(['Price'],axis=1)
y=house['Price']
```

```python
x=pd.get_dummies(x,dtype='int')#,'Regionname','CouncilArea']])
```

```python
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
model1=Lasso(alpha=100,max_iter=150,tol=0.2)
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
model2=Ridge(alpha=50,max_iter=150,tol=0.2)
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```
